```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import linear_model
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import statistics


# In[2]:


df=pd.read_csv(r'C:\Users\user\Desktop\HepatitisCdata.csv')


# In[3]:


print("Our project shape is: ")
df.shape


# In[4]:


df


# ## CLEANING AND PREPARING DATA

# In[5]:


df.isnull().sum()


# In[6]:


missing_value=["N/a","na",np.nan]


# In[7]:


df=pd.read_csv(r'C:\Users\user\Desktop\HepatitisCdata.csv',na_values=missing_value)


# In[8]:


df.isnull().sum()


# In[9]:


sns.heatmap(df.isnull(),yticklabels=False)
plt.title("Graph showing where are null values present")
plt.show()


# In[10]:


df2=df.fillna(value=0)


# In[11]:


df2.isnull().sum()


# In[12]:


df2.set_index("Index")


# In[13]:


df2=df2.dropna()


# In[14]:


df2['Age'] = df2['Age'].astype(int)


# In[15]:


df2['Category'].replace({'0=Blood Donor' : 0, '0s=suspect Blood Donor' : 1,  '1=Hepatitis' : 2, '2=Fibrosis' : 3, '3=Cirrhosis' : 4}, inplace = True)


# In[16]:


df2['Sex'].replace({'m' : 0, 'f' : 1}, inplace = True)


# In[17]:


df2


# In[18]:


print(df2.dtypes)


# ## STATISTICAL DATA AND GRAPHS

# In[19]:


from scipy.stats import pearsonr
print(df2.describe())


# #### STATISTICAL DATA FOR AGE

# In[20]:


print("Mean:   ",df2['Age'].mean())
print("Median: ",df2['Age'].median())
print("Mode: ",statistics.mode(df2['Age']))
print("Standard deviation: ",df2['Age'].std())
print("Variance: ",df2['Age'].var())
print("Minimum value: ",df2['Age'].min())
print("Maximum value: ",df2['Age'].max())


# #### STATISTICAL DATA FOR CHOLINESTERASE

# In[21]:


print("Mean:   ",df2['CHE'].mean())
print("Median: ",df2['CHE'].median())
print("Mode: ",statistics.mode(df2['CHE']))
print("Standard deviation: ",df2['CHE'].std())
print("Variance: ",df2['CHE'].var())
print("Minimum value: ",df2['CHE'].min())
print("Maximum value: ",df2['CHE'].max())


# #### STATISTICAL DATA FOR CHOLESTROL

# In[22]:


print("Mean:   ",df2['CHOL'].mean())
print("Median: ",df2['CHOL'].median())
print("Mode: ",statistics.mode(df2['CHOL']))
print("Standard deviation: ",df2['CHOL'].std())
print("Variance: ",df2['CHOL'].var())
print("Minimum value: ",df2['CHOL'].min())
print("Maximum value: ",df2['CHOL'].max())


# In[23]:


df2.groupby(by=['Category','Sex']).mean()


# In[60]:


plt.figure(figsize=(15, 8))
types = ['0=Blood Donor','0s=suspect Blood Donor','1=Hepatitis', '2=Fibrosis', '3=Cirrhosis']
cols = ['c','m','r','b','y']
plt.pie(df2.groupby('Category').size(),colors=cols,startangle=90,explode=(0,0.5,0.5,0.5,0.5),autopct='%0.1f%%')
plt.legend(types)
plt.title("Category division")
plt.show()


# In[25]:


plt.figure(figsize=(10, 5))
df2['Sex'].replace({0 : 'male', 1 : 'female'}, inplace = True)
sns.countplot(x ='Sex', data = df2)
df2['Sex'].replace({'male' : 0, 'female' : 1}, inplace = True)
plt.title("Count of MALES and FEMALES")
plt.show()


# In[26]:


CHE = df2['CHE']
CHOL = df2['CHOL']
columns1 = [CHE,CHOL]
fig, ax = plt.subplots()
ax.boxplot(columns1)
ax.set_title("Distribution of Cholinesterase levels and Cholestrol levels")
plt.xlabel("CHE(1) and CHOL(2)")
plt.ylabel("values")
plt.show()


# In[27]:


plt.figure(figsize=(15, 5))
sns.countplot(df2['Age'],hue=df2['Sex'],palette='copper')
plt.ylabel("Count of each Sex")
plt.xlabel("Age")
plt.title("How does count of Sex varies with age")
plt.show()


# In[28]:


plt.figure(figsize=(20, 10))
plt.plot(list(df2[df2['Sex'].isin([0])]['CHOL']), "-", color='blue')
plt.plot(list(df2[df2['Sex'].isin([1])]['CHOL']), "-", color='red')
plt.xlabel("Number of People")
plt.ylabel("Cholestrol levels")
plt.title("Comparision in Males and Females and their Cholestrol level")
plt.legend(['male','female'])
plt.show()


# In[29]:


plt.figure(figsize=(20, 5))
CHOL = df2['CHOL']
Age = df2['Age']
plt.scatter(Age,CHOL,color='r')
plt.xlabel('Age')
plt.ylabel('Cholestrol level')
plt.title("How Age affects cholestrol levels")
plt.show()


# ## SKLEARN  LIBRARY MACHINE ALGORITHM

# In[40]:


df2.shape


# In[41]:


df2.groupby('Category').size()


# In[42]:


array = df2.values
X = array[:,0:11]
Y = array[:,10]


# In[43]:


t_size = 0.40
seed = 7
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=t_size,random_state=seed)
print("X_train: ", X_train.shape)
print("Y_train : ", Y_train.shape)
print("X_test : ", X_test.shape)
print("Y_test : ", Y_test.shape)


# In[44]:


print(df2.dtypes)
df2['Index'] = df2['Index'].astype(int)
df2['Category'] = df2['Category'].astype(int)
df2['Sex'] = df2['Sex'].astype(int)
df2['ALB'] = df2['ALB'].astype(int)
df2['ALP'] = df2['ALP'].astype(int)
df2['ALT'] = df2['ALT'].astype(int)
df2['AST'] = df2['AST'].astype(int)
df2['BIL'] = df2['BIL'].astype(int)
df2['CHE'] = df2['CHE'].astype(int)
df2['CHOL'] = df2['CHOL'].astype(int)
df2['CREA'] = df2['CREA'].astype(int)
df2['GGT'] = df2['GGT'].astype(int)
df2['PROT'] = df2['PROT'].astype(int)


# In[45]:


knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train, Y_train)
predictions = knn.predict(X_test)
print((accuracy_score(Y_test, predictions)))


# In[46]:


print((classification_report(Y_test, predictions)))
print(confusion_matrix(Y_test, predictions))


# In[47]:


dt = DecisionTreeClassifier( criterion='entropy', ccp_alpha = 0.015)
dt.fit(X_train, Y_train)
predictions = dt.predict(X_test)
print((accuracy_score(Y_test, predictions)))


# In[48]:


dt = DecisionTreeClassifier( criterion='gini', ccp_alpha = 0.015)
dt.fit(X_train, Y_train)
predictions = dt.predict(X_test)
print((accuracy_score(Y_test, predictions)))


# In[49]:


nb = GaussianNB()
nb.fit(X_train, Y_train)
predictions = nb.predict(X_test)
print((accuracy_score(Y_test, predictions)))


# In[50]:


X = df2[['Sex','ALB','ALP','ALT','AST','BIL','CHE','CHOL','CREA','GGT','PROT']]
Y = df2['Category']


# In[51]:


regr = linear_model.LinearRegression()
regr.fit(X, Y)
print(regr.coef_)


# In[52]:


predictedHD = regr.predict([[0,40,50,20,20,8,7,4,80,20,65]])
print(predictedHD)


# In[53]:


scaler = StandardScaler()
data_scaled = scaler.fit_transform(df2)
kmeans = KMeans(n_clusters=2, init='k-means++')
kmeans.fit(data_scaled)


# In[54]:


kmeans.inertia_


# In[55]:


SSE = []
for cluster in range(1,20):
    kmeans = KMeans(n_clusters = cluster, init='k-means++')
    kmeans.fit(data_scaled)
    SSE.append(kmeans.inertia_)

frame = pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE})
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()


# In[56]:


kmeans = KMeans(n_clusters = 5, init='k-means++')
kmeans.fit(data_scaled)
pred = kmeans.predict(data_scaled)
frame = pd.DataFrame(data_scaled)
frame['cluster'] = pred
frame['cluster'].value_counts()


# ## END OF PROJECT
```
